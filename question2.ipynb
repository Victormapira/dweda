```python
import os
import shutil
from PIL import Image

original_folder = ".\\ANIMAL DATASET"
merged_image_dataset = ".\\MERGED_DATASET"

for item in os.listdir(original_folder):
	src_file = os.path.join(original_folder,item)
	if os.path.isfile(src_file):
		shutil.copy2(src_file,merged_image_dataset)
	elif os.path.isdir(src_file):
		for sub_items in os.listdir(src_file):
			sub_src_file = os.path.join(src_file,sub_items)
			shutil.copy2(sub_src_file,merged_image_dataset) 

resized_folder = ".\\MERGED_DATASET_RESIZED"
target_size = (230,230)

if not os.path.exists(resized_folder):
    os.makedirs(resized_folder)
total_images=1
#itarate over each image in the original folder 
for image_files in os.listdir(merged_image_dataset):
    if image_files.endswith((".jpg",".png",".jpeg",".gif")):
        total_images +=1
        with Image.open(os.path.join(merged_image_dataset,image_files)) as img:
            resized_img=img.resize(target_size,Image.LANCZOS)
            #save the resized image to the resized folder
            resized_img.save(os.path.join(resized_folder,image_files))
            print(f"Resized {image_files} succesfully")
```
